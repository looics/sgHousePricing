```python
def checkUniqueColumns(df1,df2,column):
    df1_list = df1[column].unique().tolist()
    df2_list = df2[column].unique().tolist()
    x = list(set(df1_list) - set(df2_list))
    return x
```

```python
print(checkUniqueColumns(df,resale_2017,'town'))
print(checkUniqueColumns(df,resale_2017,'street_name'))
```
